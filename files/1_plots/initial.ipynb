````python
gtf = open("/import/48", 'r')
````

````python
gtf.seek(0)

mapped = {}

#limit = 20
for line in gtf:
    if line[0] != '#':

        #limit = limit - 1
        #if limit < 0:
        #    break
        
        j,j,j,j,j,j,j,j,stuff = line.split('\t')
        iterable = (p.strip().split(' ') for p in stuff.split(';'))
        
        gene_id = None; gene_name = None; exon_id = None; 
        exon_number = None; transcript_name = None
        
        for a in iterable:
            if len(a)!=2:
                continue
            x = a[0]
            y = a[1]
            if x=="exon_id":
                exon_id = y.split('"')[1]
            elif x=="exon_number":
                exon_number = int(y)
            elif x=="gene_id":
                gene_id = y.split('"')[1]
            elif x=="gene_name":
                gene_name = y.split('"')[1]
            elif x=="transcript_name":
                transcript_name = y.split('"')[1]
                
        if exon_id:
            if exon_id not in mapped:
                # insert new genes
                mapped[exon_id] = [gene_name, {transcript_name : [exon_number]}]
            else:
                if gene_name != mapped[exon_id][0]:
                    print("Different Gene:", exon_id, mapped[exon_id][0], gene_name)
                else:
                    # Same gene, different transcript?
                    trans_names = mapped[exon_id][1]
                    if transcript_name not in trans_names:
                        # insert trans exon number
                        mapped[exon_id][1][transcript_name] = [exon_number]
                    else:
                        # Same transcript, different exon_number?
                        if exon_number not in mapped[exon_id][1][transcript_name]:
                            mapped[exon_id][1][transcript_name].append(exon_number)
````

````python
matrices = {}
for imp in sorted(import_map.keys(), key=int):
    mat_name = import_map[imp]
    matrices[mat_name] = open("/import/" + imp, 'r')
    print(mat_name, end="\t")
````

````python
def parseSingleMatrix(name, matrix):
    matrix.seek(0)
    fope=open("%s_fixednames.tsv" % name, "w")
    print(fope.name, end=", ")
    print('\t'.join(matrix.readline().replace("\t","\t%s:" % name)), file=fope, end="")
    for mat in matrix:
        tokens = mat.split('\t')
        name = tokens[0]
        ename = mapped[name]

        # Flatten names
        flatten = [ename[0]]
        for tname in ename[1]:
            enums = ename[1][tname]
            for e in enums:
                flatten.append(tname+"_"+str(e))
        flatten.append(name)

        tokens[0] = '|'.join(flatten)
        print('\t'.join(tokens), file=fope, end="")
    fope.close()
````

````python
for name in matrices:
    title = name.split('_')[1].split('.')[0]
    parseSingleMatrix(title, matrices[name])
    print("done.", end="\t")
````

````python
from pandas import DataFrame
````

````python
names = list(map(lambda x: x.split('_')[1].split('.')[0]+'_fixednames.tsv', sorted(matrices.keys())))
#names
````

````python
#Headers
fopen.close()
fopen = open("results.tsv", "w")
print('matrix', 'gene', 'exon', 'ens', 'ncells', 'ntrans', sep='\t')


def dot1lscore(matrix):
    fore = open(matrix, "r")
    nam = matrix.split('_')[0]
    ASD = []
    for line in fore:
        if line.startswith("Dot1"):
            tmp = line.split('\t')
            num_cells = 0
            total_transcripts = 0
            transname = tmp[0]
            for cell in tmp[1:]:
                c = int(cell)
                if c > 0:
                    num_cells += 1
                    total_transcripts += c
            try:
                gene, exon, ens = transname.split('|')
            except ValueError:
                tokens = transname.split('|')
                gene = tokens[0]
                exon = '|'.join(tokens[1:-2])
                ens = tokens[-1]
            ASD.append([nam, gene, exon, ens, num_cells, total_transcripts])
    fore.close()
    return(ASD)
````

````python
#dot1lscore(names[0])
data = DataFrame(columns=['matrix','gene','exon','ensmuse', 'ncells', 'ntrans'])
for nam in names:
    scores = dot1lscore(nam)
    df1 = DataFrame(scores, columns=['matrix','gene','exon','ensmuse', 'ncells', 'ntrans'])
    data = data.append(df1, ignore_index=True)
data
````
